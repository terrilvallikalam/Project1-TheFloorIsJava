```python
# Import Dependencies
import pandas as pd
```

```python
# Make a reference to the netflix_titles.csv file path
csv_path = "Resources/netflix_titles.csv"

# Import the netflix_titles.csv file as a DataFrame
netflix_df = pd.read_csv(csv_path, encoding="utf-8")
netflix_df.head()
```

```python
# Extract relevant columns
new_netflix_df = netflix_df.loc[:, ['type', 'title', 'country', 'date_added',
       'release_year', 'rating', 'duration', 'listed_in']]
new_netflix_df
```

```python
# Initial count in columns
new_netflix_df.count()
```

```python
#Drop columns with NaN
netflix = new_netflix_df.dropna()
```

```python
#Final count in columns
netflix.count()
```

```python
#Show new Data Frame
netflix
```

```python
#Rename columns
netflix_renamed = netflix.rename(columns={"type": "Movie/TV Show", 
                                          "title":"Title", 
                                          "country":"Country of Production", 
                                          "date_added":"Date Added to Netflix", 
                                          "release_year":"Release Year", 
                                          "rating":"Parental Rating", 
                                          "duration":"Duration",
                                          "listed_in":"Genre"})
netflix_renamed
```

```python
# Split "Date Added to Netflix"
year_added = netflix_renamed['Date Added to Netflix'].str.split(',', expand=True)
year_added
```

```python
# Split "Country of Production"
countries = netflix_renamed['Country of Production'].str.split(',', expand=True)
countries
```

```python
# Rename country columns
countries_renamed = countries.rename(columns={0: "Country 1", 
                                          1:"Country 2", 
                                          2:"Country 3", 
                                          3:"Country 4", 
                                          4:"Country 5", 
                                          5:"Country 6", 
                                          6:"Country 7", 
                                          7:"Country 8", 
                                          8:"Country 9",
                                          9:"Country 10",
                                           10:"Country 11",
                                           11:"Country 12"})
countries_renamed
```

```python
# Renamed year_added
year_added = year_added.rename(columns={0: "Calendar Date Added to Netflix", 
                                          1:"Year Added to Netflix"})
```

```python
# Join netflix_final and countries_renamed dataframes
netflix_final = netflix_renamed.join([year_added, countries_renamed])
netflix_final
```

```python
# Converting "Year Added to Netflix to integer
netflix_final["Year Added to Netflix"] = pd.to_numeric(netflix_final["Year Added to Netflix"])
netflix_final["Year Added to Netflix"] = netflix_final["Year Added to Netflix"].astype(int)
netflix_final.dtypes
```

```python
netflix_final["Country of Production"] = netflix_final["Country of Production"].str.replace(", ", ",")
```

```python
netflix_final["Country of Production"] = netflix_final["Country of Production"].str.strip()
```
