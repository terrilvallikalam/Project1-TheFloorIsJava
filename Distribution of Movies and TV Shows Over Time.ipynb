Distribution of Movies and TV Shows Over Time

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# Import the netflix_titles.csv file as a DataFrame
netflix_df = pd.read_csv("Resources/clean_netflix.csv", encoding="utf-8")
netflix_df
```

```python
# Number of Movies vs. TV Shows
netflix_df['Movie/TV Show'].value_counts()
```

```python
labels = ["Movies", "TV Shows"]
sizes = [5377, 2410]
colors = ["red", "cyan"]
plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", startangle=0)
plt.title("Percentage of Movies and TV Shows on Netflix")
plt.show()
```

```python
# GroupBy
netflix_year = netflix_df.groupby(["Year Added to Netflix"])["Movie/TV Show"].value_counts()
netflix_year
```

```python
# Bin Movies/TV Shows by Years
netflix_year.plot.bar(x="Year Added to Netflix", y="Movie/TV Show", stacked=True)
```

```python
netflix_df.plot.bar(x="Year Added to Netflix", y="Release Year", stacked=True)
```
